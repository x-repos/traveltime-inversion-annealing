Custom Tikhonov Regularization Method with Initial Guess

```python
from dwave.system import DWaveSampler, EmbeddingComposite # type: ignore
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
import sys, os, importlib

# Reload modules to ensure the latest changes are picked up
sys.path.append(os.path.abspath("modules"))
import modules.plotting
import modules.dwaveapi
import modules.gendata
importlib.reload(modules.plotting)
importlib.reload(modules.dwaveapi)
importlib.reload(modules.gendata)

from modules.dwaveapi import *
from modules.gendata import *
from modules.plotting import *

np.random.seed(99) # random noise
```

```python
T = np.load("./input/T-noise-1.npy")
D = np.load('./input/D.npy')
rows, cols = 30, 10

s0 = np.ones((rows, cols))*1/3500
s0 = s0.flatten()
# s0 = s1.copy()

# Regularization parameters
lambda_reg = 2  # Regularization strength

# Define the regularized least squares with initial guess
I = np.eye(D.shape[1])  # Identity matrix for regularization
ATA = D.T @ D
rhs = D.T @ T + lambda_reg * s0  # Regularization pulls the solution towards s1
s = np.linalg.solve(ATA + lambda_reg * I, rhs)

v_inversion = (1.0 / s).reshape(rows, cols)

plt.imshow(v_inversion, cmap='viridis')
plt.colorbar(label='Inverted Velocity (m/s)')
plt.title('Velocity Inversion')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
np.savetxt("./results-tikhonov/inv-noise-1.txt", v_inversion)
```
